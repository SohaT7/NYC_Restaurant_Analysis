```python
# Import dependencies and load in CSV files
import pandas as pd
import numpy as np
from pathlib import Path 

joined_database_df = pd.read_csv("FINAL_NYC_restaurants_full_database.csv") 
                            
```

```python
joined_database_df
```

```python
# Rename to all capital for consistent column names
joined_database_df.rename(columns = {'STREET':'street','INCOME_LEVEL':'income_level', 'CUISINE_DESCRIPTION':'cuisine_description', 'BOROUGH':'borough', 'GRADE':'grade', 'SCORE':'score', 'ZIPCODE':'zipcode' }, inplace = True)
```

```python
joined_database_df
```

```python
joined_database_df=joined_database_df[['zipcode','DBA','street','income_level','borough', 'cuisine_description','score','grade', ]]
```

```python
joined_database_df=joined_database_df.sort_values(by='zipcode',ascending=True)
joined_database_df.head(10)
```
